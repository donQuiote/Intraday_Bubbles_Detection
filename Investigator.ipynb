```python
df = df.with_columns(
    pl.col("trade-price").pct_change().alias("percent_change")
) 
```

```python
df = df.with_columns(
    (pl.when(pl.col("percent_change") <= 0).then(-1).otherwise(+1)).alias("signal"),
    (pl.col("trade-volume").pct_change()).alias("vol_change")
)
```

```python
time_series = df.select(["signal","vol_change"]).with_columns(
    (pl.col("signal")-pl.col("signal").mean())/pl.col("signal").std(ddof=1),
    (pl.col("vol_change")-pl.col("vol_change").mean())/pl.col("vol_change").std(ddof=1)
)
time_series.max()
```

```python
signal = time_series.select("signal").to_pandas()["signal"]
vol_change = time_series.select("vol_change").to_pandas()["vol_change"]

# Create a plot
plt.figure(figsize=(12, 6))
plt.plot(signal, label='Signal (Normalized)', color='blue', alpha=0.7)
plt.plot(vol_change, label='Vol Change (Normalized)', color='orange', alpha=0.7)

# Add labels, legend, and title
plt.title("Time Series Comparison", fontsize=16)
plt.xlabel("Time Index", fontsize=14)
plt.ylabel("Normalized Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
correlation = time_series.select(
    pl.corr("signal", "vol_change")
).to_numpy()[0]  # Extract the result as a scalar
print(f"Correlation between 'signal' and 'vol_change': {correlation}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Extract the `trade-volume` column from Polars
trade_volume = df.select("trade-volume").to_pandas()["trade-volume"]

# Plot the PDF using seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(trade_volume, fill=True, color='blue', alpha=0.5)

# Add titles and labels
plt.title("PDF of Trade Volume", fontsize=16)
plt.xlabel("Trade Volume", fontsize=14)
plt.ylabel("Density", fontsize=14)

# Show the plot
plt.grid(True)
plt.show()
```
